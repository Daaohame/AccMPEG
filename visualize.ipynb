```python
import matplotlib.pyplot as plt
```

```python
mpeg = []
compressed1 = []
compressed2 = []
with open('stats', 'r') as f:
    for line in f.readlines():
        line = line.split(',')
        name = line[0]
        acc = float(line[1])
        bw = int(line[2])
        
        if 'compressed' in name:
            if acc < 0.97:
                continue
            if 'base_34' in name and 'cont_weight_0' in name:
                compressed1.append({
                    'acc': acc,
                    'bw': bw,
                    'name': name
                }) 
            if 'base_34' in name and 'cont_weight_32' in name:
                compressed2.append({
                    'acc': acc,
                    'bw': bw,
                    'name': name
                }) 
            # compressed1.append({
            #     'acc': acc,
            #     'bw': bw,
            #     'name': name
            # })
        else:
            mpeg.append({
                'acc': acc,
                'bw': bw
            })
```

```python
plt.style.use('ggplot')
```

```python
plt.scatter([i['bw'] for i in mpeg], [i['acc'] for i in mpeg])

plt.scatter([i['bw'] for i in compressed1], [i['acc'] for i in compressed1])
plt.scatter([i['bw'] for i in compressed2], [i['acc'] for i in compressed2])

plt.xlim(left=0)
plt.xlabel('Bandwidth (B)')
plt.ylabel('Accuracy (F1)')
```
